```python
sn.countplot(x= 'Gender', data=df)
```

```python
ax = sn.countplot(x='Gender', data=df)

for bars in ax.containers: 
    ax.bar_label(bars)
    
```

The graph showing Female count is more than male count 

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sales_gen =  df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sn.barplot(x='Gender', y='Amount', data=sales_gen)
```

Age

```python
ax  = sn.countplot(data = df, x ="Age Group", hue = "Gender")

for bars in ax.containers:
    ax.bar_label(bars) 
```

The graph showing which age group has more count

```python
sales_age =  df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sn.barplot(x='Age Group', y='Amount', data=sales_age)
```

state


```python
sales_state =  df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)#first 10 state
sn.set(rc={'figure.figsize':(9,5)})
sn.barplot(x='State', y='Orders', data=sales_state) 
```

total numbers of orders state vise
most of the orders are uttar pradesh,Maharashtra and karnatka

Marital Status

```python
ax = sn.countplot(data = df, x='Marital_Status')
sn.set(rc={'figure.figsize':(7,9)})#width and height
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_stat =  df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sn.set(rc={'figure.figsize':(6,5)})
sn.barplot(x='Marital_Status', y='Amount', data=sales_stat, hue='Gender')
```

Here we can understand, marriaged female done more purchase

Occupation

```python
sn.set(rc={'figure.figsize':(9,10)})
ax = sn.countplot(data = df, x='Occupation')
for bars in ax.containers:
    ax.bar_label(bars)

```

It section job category purchased more

```python
sales_stat =  df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sn.set(rc={'figure.figsize':(9,10)})
sn.barplot(x='Occupation', y='Amount', data=sales_stat)
```

"From the above graph we can see that most of the buyers are working in IT, Healthcare,aviation and sector*

Product Category

```python
sn.set(rc={'figure.figsize':(16,6)})
ax = sn.countplot(data = df, x='Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)

```

High Sales done based the product_categor

```python
sales_stat =  df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head()
sn.set(rc={'figure.figsize':(15,6)})
sn.barplot(x='Product_Category', y='Amount', data=sales_stat)
```

From the above graph we can see that most of the sold products are from Food,Footwear and Electronics

```python
sales_stat =  df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head()
sn.set(rc={'figure.figsize':(15,6)})
sn.barplot(x='Product_ID', y='Orders', data=sales_stat)
```

Conclusion

Married woemn age group 26-35 yrs fromUP,Maharashtra and Karntaka working in IT,Healthcare and Aviation are more likely to buy products from Food Clothing and Electronics Category

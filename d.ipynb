```python
import numpy as np 
import pandas as pd 
import seaborn as sn
import matplotlib.pyplot as plt  


```

```python
df = pd.read_csv("ds_salaries.csv")
```

```python
df.drop(['demos', 'Unnamed: 12'], axis = 1, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.isnull().sum() 
```

```python
# Convert categorical variables to 'category' dtype
df['employee_residence'] = df['employee_residence'].astype('category')
df['company_location'] = df['company_location'].astype('category')
df['company_size'] = df['company_size'].astype('category')

# Map remote_ratio to readable labels if needed
df['remote_ratio'] = df['remote_ratio'].map({0: 'No remote', 50: 'Partially remote', 100: 'Fully remote'})#numerical to categorical values
df['remote_ratio'] = df['remote_ratio'].astype('category')
```

```python
from sklearn.preprocessing import StandardScaler

# Standardize salary_in_usd
scaler = StandardScaler()
df['salary_in_usd'] = scaler.fit_transform(df[['salary_in_usd']] ) #range between 0 and 1(mean of $50,000)
```

```python
import matplotlib.pyplot as plt

# Visualize distribution of salaries
plt.figure(figsize=(10, 6))
sn.histplot(df['salary_in_usd'], kde=True)
plt.title('Distribution of Salaries in USD')
plt.xlabel('Salary (USD)')
plt.ylabel('Frequency')
plt.show()
```

Salary value between -1 and  0 have higher frequency(count)

```python
plt.figure(figsize=(8, 5))
ax=sn.countplot(x='remote_ratio', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Distribution of Remote Work')
plt.xlabel('Remote Work Ratio')
plt.ylabel('Count')
plt.show()
```

Fully remote has higher count than no remote work ratio and partially remote

```python
plt.figure(figsize=(10, 6))
sn.boxplot(x='company_size', y='salary_in_usd', data=df)
plt.title('Salary Distribution by Company Size')
plt.xlabel('Company Size')
plt.ylabel('Salary (USD)')
plt.show()
```

Small-scale industries have a broader salary distribution range compared to large and medium-sized industries

```python
plt.figure(figsize=(12, 8))
sn.boxplot(x='job_title', y='salary_in_usd', data=df)
plt.title('Salary Distribution by Job Title')
plt.xlabel('Job Title')
plt.ylabel('Salary (USD)')
plt.xticks(rotation=90)  # Rotate job titles if they overlap
plt.show()
```

Research Scientist got higher salary range(USD)

```python
plt.figure(figsize=(10, 6))
ax = sn.countplot(x='company_size', data=df, palette='Set2')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Count of Jobs by Company Size')
plt.xlabel('Company Size')
plt.ylabel('Number of Jobs')
plt.show()
```
